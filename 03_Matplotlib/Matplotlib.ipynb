# Histogram plot :-  Used to visualize the distribution of a single continuous variable. Useful for understanding the frequency or probability distribution of data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.randn(1000)

# Create histogram
plt.hist(data, bins=30, color='red', edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

# Show plot
plt.show()
```

# Pie charts:-  Used to show the proportion of different categories within a dataset. Useful for representing parts of a whole.

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 20, 25, 25]  # Percentage distribution
colors = ['red', 'yellow', 'orange', 'green']
explode = (0, 0, 0.1, 0)  # Explode the 3rd slice (Oranges)

# Plot
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Fruit Distribution')

plt.show()
```

# Box plots:- Used to visualize the distribution of a continuous variable and to identify outliers or variability in the data. Useful for comparing distributions across different groups or categories.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating some random data
np.random.seed(10)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Creating a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, patch_artist=True)

# Adding labels and title
plt.xticks([1, 2, 3], ['Data 1', 'Data 2', 'Data 3'])
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.title('Box Plot')

plt.grid(True)

# Showing the plot
plt.show()
```

# Contour plots

- ***Contour plots are particularly useful for visualizing three-dimensional data in two dimensions. They display the level curves of a function of two variables on a plane. These plots are commonly used in fields such as geography, meteorology, physics, and engineering to represent topographical or other spatial data.***

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

# Create contour plot
plt.contour(X, Y, Z, cmap='viridis')
plt.colorbar()  # Add colorbar for reference

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Contour Plot')

# Show plot
plt.show()
```

# Heatmaps
# Heatmaps are graphical representations of data where individual values contained in a matrix are represented as colors. They are commonly used in data analysis and visualization to visualize the intensity of data at different points, such as the correlation between variables in a correlation matrix or the frequency of occurrences in a two-dimensional dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.rand(10, 10)

# Create heatmap
plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar()  # Add colorbar for reference

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap')

# Show plot
plt.show()
```

# 3D Plot

- **Creating 3D plots in Matplotlib allows you to visualize data in three dimensions, which is useful for representing complex spatial relationships or functions of two variables. Here's an example of creating a simple 3D plot:**

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('3D Plot')

# Show plot
plt.show()
```

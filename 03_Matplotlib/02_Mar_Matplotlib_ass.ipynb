# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
- import numpy as np
- np.random.seed(3)
- x = 3 + np.random.normal(0, 2, 50)
- y = 3 + np.random.normal(0, 2, len(x))
- Note: Also add title, xlabel, and ylabel to the plot.

- **A scatter plot is a type of plot that visualizes the relationship between two continuous variables. Each point on the plot represents a single observation, with its position determined by the values of the two variables being compared.**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create scatter plot
plt.scatter(x, y, c="red", s=200, alpha=0.6 , marker="v")

# Add title and labels
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
- import numpy as np
- For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
- For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
- For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
- For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

- ***The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure. It allows you to arrange plots in a grid-like structure, specifying the number of rows and columns and the position of each subplot within the grid.***

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Define y values for each line plot
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots and draw line plots
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, subplot position 1
plt.plot(x, y1)
plt.title('Line Plot 1')

# Subplot 2
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, subplot position 2
plt.plot(x, y2)
plt.title('Line Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, subplot position 3
plt.plot(x, y3)
plt.title('Line Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, subplot position 4
plt.plot(x, y4)
plt.title('Line Plot 4')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. import numpy as np
- company = np.array(["Apple", "Microsoft", "Google", "AMD"])
- profit = np.array([3000, 8000, 1000, 10000])

# A bar plot is a type of plot that visualizes categorical data with rectangular bars. The height or length of each bar represents the value of the corresponding category. Bar plots are commonly used to compare the values of different categories or to show changes in a single category over time.

- Bar plots are used for the following purposes:

- Comparison: Bar plots are useful for comparing the values of different categories. They make it easy to see which category has the highest or lowest value.

- Trends: Bar plots can show changes in a single category over time. By plotting data for different time periods, you can visualize trends and patterns.

- Distribution: Bar plots can also be used to visualize the distribution of categorical data. They provide a visual representation of the frequency or proportion of each category.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.title('Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show plot
plt.show()

# Create horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightgreen')
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit ($)')
plt.ylabel('Company')

# Show plot
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
- box1 = np.random.normal(100, 10, 200)
- box2 = np.random.normal(90, 20, 200)

# A box plot, also known as a box-and-whisker plot, is a method for graphically depicting groups of numerical data through their quartiles. It provides a summary of the distribution of a dataset, including the minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum values. Additionally, it can display outliers in the data.

- Box plots are used for the following purposes:

- Visualizing distribution: Box plots provide a visual summary of the distribution of numerical data, making it easy to understand the central tendency, spread, and skewness of the dataset. 

- Identifying outliers: Box plots visually highlight outliers in the dataset, which are data points that lie significantly far from the rest of the data.

- Comparing distributions: Box plots allow for the comparison of the distributions of multiple datasets or groups, making it easy to identify differences in central tendency, spread, and variability.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])

# Add title and labels
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Value')

# Show plot
plt.show()
```

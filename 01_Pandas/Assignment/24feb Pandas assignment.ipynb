## Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
import pandas as pd

# Creating a sample DataFrame
df = pd.DataFrame({'A': [10, 20, 30, 40],
                   'B': [50, 60, 70, 80],
                   'C': [90, 100, 110, 120]})
df
```

```python
def reindex_dataframe(df):
    # Resetting the index and dropping the current index
    df.reset_index(drop=True, inplace=True)
    
    # Generating new index values starting from 1 and incrementing by 2
    new_index = [(i * 2) + 1 for i in range(len(df))]
    
    # Setting the new index to the DataFrame
    df.index = new_index
    
    return df

# Applying the function to the DataFrame
df_reindexed = reindex_dataframe(df)
df_reindexed
```

## Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console.

```python
import pandas as pd

# Creating a DataFrame with a 'Values' column
data = {'Values': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

df
```

```python
def sum_first_three_values(df):
    # Check if the 'Values' column exists in the DataFrame
    if 'Values' not in df.columns:
        print("Error: 'Values' column not found in the DataFrame.")
        return
    
    # Get the first three values from the 'Values' column
    first_three_values = df['Values'].head(3)
    
    # Calculate the sum of the first three values
    sum_of_first_three = first_three_values.sum()
    
    # Print the sum to the console
    print("Sum of the first three values:", sum_of_first_three)
```

```python
# Applying the function to the DataFrame
sum_first_three_values(df)
```

## Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
import pandas as pd

# Creating a DataFrame with a 'Text' column
data = {'Text': ["This is a sample text", "Count the words in this sentence", "Another example sentence"]}
df = pd.DataFrame(data)
df
```

```python
def count_words(text):
    # Split the text into words using whitespace as delimiter
    words = text.split()
    # Count the number of words
    word_count = len(words)
    return word_count

def add_word_count_column(df):
    # Check if the 'Text' column exists in the DataFrame
    if 'Text' not in df.columns:
        print("Error: 'Text' column not found in the DataFrame.")
        return
    
    # Apply the count_words function to each row of the 'Text' column
    df['Word_Count'] = df['Text'].apply(count_words)

    return df
```

```python
# Applying the function to the DataFrame
df_with_word_count = add_word_count_column(df)
print("DataFrame with Word_Count column:")
df_with_word_count
```

## Q5. How are DataFrame.size() and DataFrame.shape() different?

- **DataFrame.size: This attribute returns the number of elements in the DataFrame, which is equal to the product of the number of rows and columns. It counts all the elements, including missing values (NaNs). The size attribute returns an integer value.Example:**

```python
import pandas as pd

# Creating a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

print("DataFrame size:", df.size)
```

- **DataFrame.shape: This attribute returns a tuple representing the dimensions of the DataFrame, i.e., the number of rows and columns. It returns the number of rows as the first element and the number of columns as the second element. The shape attribute returns a tuple.Example:**

```python
import pandas as pd

# Creating a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

print("DataFrame shape:", df.shape)
```

- ***In summary***:

- **DataFrame.size returns the total number of elements in the DataFrame.**
- **DataFrame.shape returns a tuple representing the dimensions (number of rows and columns) of the DataFrame.**

## Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address. 

## The username is the part of the email address that appears before the '@' symbol. For example, if the email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your function should extract the username from each email address and store it in the new 'Username' column.

```python
import pandas as pd

# Sample DataFrame with email addresses
df = pd.DataFrame({'Email': ['john.doe@example.com', 'alice.smith@example.com', 'bob.johnson@example.com']})
df
```

```python
def extract_username(email):
    # Split the email address using '@' as delimiter and return the first part
    return email.split('@')[0]

def add_username_column(df):
    # Check if the 'Email' column exists in the DataFrame
    if 'Email' not in df.columns:
        print("Error: 'Email' column not found in the DataFrame.")
        return
    
    # Apply the extract_username function to each row of the 'Email' column
    df['Username'] = df['Email'].apply(extract_username)

    return df

# Applying the function to the DataFrame
df_with_username = add_username_column(df)

print("\nDataFrame with Username column:")
df_with_username
```

## Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selectsall rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The function should return a new DataFrame that contains only the selected rows.


```python
import pandas as pd

# Create DataFrame with sample data
data = {'A': [3, 8, 6, 2, 9],
        'B': [5, 2, 9, 3, 1],
        'C': [1, 7, 4, 5, 2]}
df = pd.DataFrame(data)

df
```

```python
def select_rows(df):
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    return selected_rows[['A', 'B', 'C']]

```

```python
selected_df = select_rows(df)
print("\nSelected DataFrame:")
selected_df
```

## Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.

```python
import pandas as pd

# Create DataFrame with sample data
data = {'Values': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
df
```

```python
def calculate_statistics(df):
    # Check if the 'Values' column exists in the DataFrame
    if 'Values' not in df.columns:
        print("Error: 'Values' column not found in the DataFrame.")
        return
    
    # Calculate mean, median, and standard deviation
    mean_value = df['Values'].mean()
    median_value = df['Values'].median()
    std_value = df['Values'].std()
    
    # Return calculated statistics
    return {'Mean': mean_value, 'Median': median_value, 'Standard Deviation': std_value}

statistics = calculate_statistics(df)
print("\nCalculated statistics:")
statistics
```

## Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python
import pandas as pd

# Create DataFrame with sample data
data = {'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05', '2022-01-06', '2022-01-07', '2022-01-08'],
        'Sales': [10, 20, 30, 40, 50, 60, 70, 80]}
df = pd.DataFrame(data)

df
```

```python
def calculate_moving_average(df):
    # Check if the 'Sales' and 'Date' columns exist in the DataFrame
    if 'Sales' not in df.columns or 'Date' not in df.columns:
        print("Error: 'Sales' or 'Date' column not found in the DataFrame.")
        return
    
    # Convert the 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Sort the DataFrame by 'Date'
    df = df.sort_values(by='Date')
    
    # Calculate the moving average using a window of size 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    
    return df

df_with_moving_average = calculate_moving_average(df)
print("\nDataFrame with Moving Average:")
df_with_moving_average
```

## Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.


```python
import pandas as pd

# Create DataFrame with sample data
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}
df = pd.DataFrame(data)
df
```

```python
def add_weekday_column(df):
    # Convert the 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Add a new column 'Weekday' containing the weekday name
    df['Weekday'] = df['Date'].dt.day_name()
    
    return df

# Apply the function to the DataFrame
df_with_weekday = add_weekday_column(df)

print("\nDataFrame with Weekday column:")
df_with_weekday
```

## Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
import pandas as pd

# Create DataFrame with sample data
data = {'Date': ['2023-01-01', '2023-01-15', '2023-01-31', '2023-02-05']}
df = pd.DataFrame(data)
df
```

```python
def select_rows_by_date(df):
    # Check if the 'Date' column exists in the DataFrame
    if 'Date' not in df.columns:
        print("Error: 'Date' column not found in the DataFrame.")
        return
    
    # Convert 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Generate a range of dates from '2023-01-01' to '2023-01-31'
    date_range = pd.date_range(start='2023-01-01', end='2023-01-31')
    
    # Filter rows where date is in the generated range
    selected_rows = df[df['Date'].isin(date_range)]
    
    return selected_rows

# Apply the function to the DataFrame
selected_df = select_rows_by_date(df)

print("\nSelected DataFrame:")
selected_df
```

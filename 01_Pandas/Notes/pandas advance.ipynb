 # read the concept of iloc and loc

 - **iloc** 
 
 Stands for "integer location".
 
Uses integer-based indexing to select rows and columns.

Accepts integer indices or slices (ranges) for both rows and columns.

***The syntax is df.iloc[rows, columns].***

The indices and slices provided are based on the numerical position of rows and columns, starting from 0.

 - **loc** 
 
.Stands for "location".
 
.Uses label-based indexing to select rows and columns based on their labels (index and column names).

.Accepts labels or boolean arrays for rows and labels for columns.

***.The syntax is df.loc[rows, columns]***

# summary

***.Use iloc when you want to select rows and columns by their integer positions.***

***.Use loc when you want to select rows and columns by their labels.***

Both iloc and loc are very useful for data manipulation and analysis tasks in Pandas, allowing you to access specific data points, rows, or columns within a DataFrame or Series.

```python
def cat_fare(x):
    if x < 10 :
        return "cheap"
    elif x >= 10 and x < 20 :
        return "mid"
    else :
        return "high"
```

```python
df['cat_fare'] = df['Fare'].apply(cat_fare)
```
